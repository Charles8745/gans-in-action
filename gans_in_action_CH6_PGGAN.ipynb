```python
import matplotlib.pyplot as plt
import tensorflow as tf
import tensorflow_hub as hub
import numpy as np
```

```python
# import the progressive GAN from TFHub
module = hub.load("https://tfhub.dev/google/progan-128/1")

# latent dimension that gets
latent_dim = 512

# Change the seed to get different faces.
latent_vector = tf.random.normal([1, latent_dim], seed=20)

# Uses module to generate images from the latent space by calling the default signature.
# The exact way to call a TF Hub module in TF2 can vary,
# but accessing the 'default' signature is a common pattern.
# If this doesn't work, inspecting `module.signatures` might be necessary.
interpolated_images = module.signatures['default'](latent_vector)['default']


# The output of the module should be directly usable in TensorFlow 2.x eager execution.
# We convert it to a NumPy array for matplotlib.
image_out = interpolated_images.numpy()


# The image output shape might be (1, H, W, C), reshape for display.
plt.imshow(image_out.reshape(128,128,3))
plt.show()
```
